If you want to add new elements to a list, you can call the append() function with  the list variable.

```python
colors =["red","blue", "green"]
print(colors)
print(len(colors))
colors.append("orange")
print(colors)
print(len(colors))
```

In this example, colors will be printed as "red", "blue", "green", "orange" with  "orange" as the last element. Adding a new element to the list also increases  its length by 1, so the last line will print out 4 instead of 3.
Lists also support operations like concatenation, which is just a fancy way for  saying addition that can be done by using the + operator.

```python
odd numbers = [1, 3, 5]
even numbers = [2, 4 ,6]
numbers =odd numbers + even numbers
print(numbers)
print(len(numbers))
```

The concatenation operation copies the elements from the first list variable odd_numbers  to numbers first, then copies the elements from the other list variable even_numbers one  by one. You can think of it as performing appending an element of the elements in the  even_numbers list in the odd_numbers list. At the end, the numbers variable becomes a  list variable with six elements as [1, 3, 5, 2, 4, 6]. Note that the order of the combined  lists are preserved.
Similarly as strings, lists support multiplication operation as well, that can be done by  using the * operator. The multiplication operation adds multiple copies of the same list. Try to run below code, see what happens?

```python
scores =[0]*10
print(scores)
print(len(scores))
```

[0 ]is a list with a single element, that is replicated 10 times by using the multiplication operator(*). 

The result is assigned to a list variable named scores, that contains 10 zeros. The code can be used to set initial values for items that have the same value. You can use  multplication operator to replicate a list with multiple elements. Try to execute below code, see what happens?

```python
actions =["ping","pong"] * 3
print(actions)
print(len(actions))
```
