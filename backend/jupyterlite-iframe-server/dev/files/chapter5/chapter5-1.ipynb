In the previous lesson, we have learnt how to store a single string or number to a  variable. What if there are multiple strings or numbers that need to be stored  together? Just like you can store either a single fruit or multiple fruits into the  same storage container, you can store either one or multiple strings or numbers to  one variable. In this situation, the variable is called a list variable. When you  assign values to a list variable, they need to be separated by commas and embedded  in a pair of square brackets.

#### Here are examples of a string list and a number list.

```python
fruits = ["apple", "banana", "cherry", "fig", "grape"]
print(fruits)

numbers =[1，2，3，4，5]
print(numbers)
```

#### We can see that:
- This list contains 5 items, or as computer scientists call them elements.
- The length of the list is the number of elements in the list. One easy way to get  a list's length is calling the len() function, same as what you can do with a string  variable.
- Each element has a position index associated with it, called the index.
- The index starts at 0 from the left most, then increases one by each element  and the largest index is 4.

We can use the index number to access the value of an element you want from a list. For example, fruits[0] is the element "apple".
To make things more interesting, Python also supports indexing from the right to the left.  It starts from -1 from the right most as the largest index, then decreases one by each step  and the smallest index is -5 in the above example.
Try to execute below code, and watch that fruits[2] and fruits[-3] refer to the same  element as "cherry".

```python
fruits = ["apple","banana","cherry","fig","grape"]
# It prints the length of the fruits list as 5, since there are 5 elements in the listprint(len(fruits))
# It prints out "cherry", with index 2 counted from the left most.print(fruits[2])
# This also prints out "cherry", with index -3 counted from the right most.print(fruits[-3])
```

You can also change a list's element by assigning a new value through its index  number. Changing an element's value does not change the length of a list, as there  are the same number of elements.

```python
fruits = ["apple", "banana", "cherry", "fig", "grape"]
print(fruits[2])
fruits[2]= "orange"
print(fruits)
print(len(fruits))
```
