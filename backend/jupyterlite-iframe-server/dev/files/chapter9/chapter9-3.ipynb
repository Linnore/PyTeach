In the below example, range(0, 10, 2) returns [0, 2, 4, 6, 8]. As illustrated by the diagram,  each step increases the number by 2 that ends up with 5 elements in the list in total. This  also provides another way to find even numbers for a given range.

```python
for num in range(0, 10, 2):
    print(num)
```
