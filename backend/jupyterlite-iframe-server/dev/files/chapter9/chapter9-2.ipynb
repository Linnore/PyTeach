Here is the diagram to help you understand how the iteration flow works. In each iteration,  the num variable is assigned with the value of the current iteration, that is added to the  previous sum variable then stored back to the sum variable. So the sum variable's value  keeps being updated with the sum of all numbers up to the current iteration, and it holds  the total sum after all the iterations complete.

To make range more powerful and slightly more complicated, the range() function can actually  support using 3 arguments,with a structure as range(start, stop, step).

- start: integer starting from which the sequence of integers is to be returned.
- stop: integer before which the sequence of integers is to be returned. The range of integers  end at stop  1, so the given endpoint is never part of the generated sequence.
- step: integer value which determines the increment between each integer in the sequence. A  step number can be even negative, that returns a list of numbers in the backward order.
