### Variable Operations
Only the operators corresponding to a variable's type are allowed for that variable. As  you may recall, the number 20 surrounded by double quotes is different from the number  20 without quotes, as the former is a string while the latter is a number.
A string variable only allows using the + and * operators, while a number variable can  use +, -, * , / and %.

```python
#s is a number variable, subtraction operator is allowed.
s = 20
print(s -10)
#s is a string variable, subtraction operator is NoT allowed and the code fails.
s ="20"
print(s - 10)
```

So you need to be always clear on what type a variable is, and what operators are allowed.
A variable name can not be used in code before a value is given to it. As an example, the  below code will fail because the variable called "name" has not been defined yet. So Python  does not know what value to put in for the variable called name.
