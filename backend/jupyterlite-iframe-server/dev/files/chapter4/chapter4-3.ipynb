You can assign variables with the values of other variables and Python will perform the  corresponding substitutions. For example, in the code below, we defined a variable called  name and gave it a value of "Bob". Then we created another variable called greeting which is  equal to the string "Hello " and the value of the variable name. Python will then find the  value of the variable name from line 1 and make a substitution. After performing the  substitution, Python will add together the 2 strings it found ("Hello " and "Bob") and give  greeting a value of "Hello Bob" which we can see by printing it out at line 3.

```python
name = "Bob'
greeting ="Hello" + name
print(greeting)
```

To make this even more interesting, you can manipulate a variable then assign the result back  to the same variable. When that happens, the value of the variable will be updated to be the  new value.

```python
score =100
score =score +50
print(score)
```

To extend on the above operation a bit, how can you swap values between two variables? For  example, say we have i = 10, j = 20, how to swap the values between them so that we get  i = 20, j = 10?
Python lets us do it by allowing multiple variables to be defined on the same line. So all  we need to do to swap the values of i and j is to write i, j = j, i. Just like how we were  able to assign a variable using the value of another variable, Python will find the  corresponding values for i and j, making the statement i,j = 20, 10 and assign each  variable on the left to the corresponding value of the right.

```python
i= 10
j= 20
i,j=j,i
print(i, j)
```

As said, the first two lines can be simplified as i, j = 10, 20. You can enable visualization mode in the Playground to observe how values of variables changes step by step.
