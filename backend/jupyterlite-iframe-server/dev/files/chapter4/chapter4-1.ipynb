In the previous lessons, we have learnt what strings and numbers are in Python. As you  may recall, some of the strings or numbers can be very long. You may wonder if there  is a convenient way to avoid writing the same long string or number repeatedly, which is  normally inefficient and can be error prone. For example, think about how you could  improve the below code.

```python
print ("pi is very very long, and here are the first ten digits of pi")
print(3.1415926535)
```

Python provides a smart way to store strings or numbers as variables, so you can use them  repeatedly. You can think of a variable as a storage container that stores a string or a  number for you.
You can specify a variable by giving it a name and assigning it a value. A variable's name  needs to start with an alphabetic character, and it can be as short as one character, or as  long as you like. The difference is that a short variable name (e.g. n) is convenient for  writing short code, while a longer variable name (e.g. name) can provide more meaningful  information about the information stored by the variable. Then you assign a value to a  variable by writing the variable name followed by an equal sign (=) and its value.
After you assign a value to a variable, you can use the variable name in place of where you  want the value to be. For example we can assign the value "Bob" to a variable called "name"  and we will be able to print out "Bob" when we use print(name).

```python
name = "Bob"
print(name)
```

The variable's value determines the type of variable. When you assign a string value to a  variable, it becomes a string variable.
