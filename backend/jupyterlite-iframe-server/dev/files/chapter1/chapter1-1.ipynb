### What is a Computer Program?

Before we dive into the Python programming language, let's briefly introduce what a computer program is. A computer program is a specific set of ordered steps designed to accomplish a specific goal. There are a few key factors that make up a computer program:

- **Dividable**: The specific goal can be divided into multiple steps.
- **Ordered**: The steps must be ordered correctly. If the order is changed, the program may produce a different result.
- **Determined**: Once the ordered steps are executed correctly, we should be able to determine the result.
- **Repeatable**: When a program is executed, it should produce the same result each time. This means the entire process is repeatable.
