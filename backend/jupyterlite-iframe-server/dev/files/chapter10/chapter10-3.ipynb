The last thing I want to teach you is that a function may have multiple return statements,  that can be put at any places you want. When a return statement is executed, it stops the  execution flow of the function immediately, and the rest code of the function will not be  executed.

In the below example, l is a list variable with its value assigned as ["apple", "banana",  "cherry"] and e is a string variable with its value assigned as "banana". When the for loop  iterates to the second element of the list, variable i is assigned the value of "banana" that  equals the value of variable e, so the contains() function returns the result as True immediately.  This prevents the for loop from further iterations and skips the last line of "return False".

As we mentioned in Chapter 8 For Loops, there are multiple indented code blocks here. The outer  code blocks will be executed before the inner code blocks. When an inner code block completes  its execution, it returns the execution control to the closest outer code block then repeats  the same process till all the code completes the execution.
