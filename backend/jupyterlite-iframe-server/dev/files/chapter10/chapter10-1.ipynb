We have learned a few functions in the previous lessons, for example print(), len() and range()  are Python built-in functions. And Python allows users to create custom functions. In Python, a  function is generally a group of related statements that perform a specific task. It avoids  writing the same code repeatedly and makes code reusable. A function has below structure.

There are 3 components in the structure:

- The function statement heading starts with the def keyword (a short for define), and followed  by the function name and parameters that can be used to pass arguments for calling the function.
- An indented code block that will be executed when the function is called.
- A return statement that returns the result to the caller. A function can also perform some  specific task without returning anything.

Same as variable names, function names have to start with an alphabetic character and not a number.  A function name can contain multiple words separated by underscores as necessary to improve readability.

To help you understand this more easily, you can think functions as recipes as illustrated in the photo.  In order to make a lemonade, you need to take lemon, sugar and water as your inputs. These are the  arguments you need to pass in a function. The recipe defines the steps to make a lemonade, that are the  statements your function needs to execute. Lastly after the steps are completed, a lemonade is made and  returned that is the output of your function.
