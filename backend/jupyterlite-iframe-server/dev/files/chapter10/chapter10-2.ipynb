Let's take a look at a simple function that does not have parameters, nor returns a result. The function is  named as hello(). Its code block contains only one print statement. To call the function, you write the  name of the function followed by parentheses - hello() and it will print out "Hello Kids Python". You can  call the function repeatedly to print out the string multiple times.

```python
def hello():
    print ("Hello Kids Python")

hello()
```

Here is another example function that takes one parameter without returning a result. The  even_odd() function takes one parameter as input, that is passed inside the parentheses.  When the function is called as even_odd(4), the num parameter is assigned a value of 4 before  executing any statement in the code block. Since 4 % 2 equals 0, even_odd(4) will print out  "4 is an even number". Similarly, when even_odd(3) is called, the num parameter is assigned  a value of 3, and "3 is an odd number" will be printed out.

```python
def even_odd(num):
    if num % 2 == 0:
        print(num, "is an even number")
    else:
        print(num, "is an odd number")

even_odd(4)
```

Further on, let's take a look at a function that returns a result. You can define as many  parameters as needed to a function. In this example, add() function takes two parameters  x and y. Its code block contains a very simple return statement, that returns the sum of  x and y. When add(2, 5) is called, the values of 2 and 5 are passed to the parameters  x and y accordingly. After the function execution completes, the sum of x + y which is 7  in this case is returned and assigned to the result variable. So the print function will  print 7 in the end.

```python
def add(x, y):
    return x + y

result = add(2, 5)
print(result)
```
