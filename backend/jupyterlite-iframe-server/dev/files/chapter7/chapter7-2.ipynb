All three lines are printed out in this example. The two prints in the green color  belong to the same code block. They are printed, since the condition is true.  The last print is unindented, that marks the end of the if-statement code block.  The unindented statements in the blue color all belong to the main code block,  and they will always be executed.

Here is the structure of if-else statement.

It contains 4 components:
- The if statement heading that contains a condition to be tested as True or False,  and ended with a colon sign (:).
- An indented code block that will be executed when the condition is True.
- The else statement heading that ends with a colon sign.
- An indented code block that will be executed when the condition is False.
