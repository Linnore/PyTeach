Let's take a closer look at how the iteration process is done with a list example

```python
fruits = ["apple", "berry", "cherry"]
for fruit in fruits:
    print(fruit)
```

n this example, the fruits variable has three elements, indexed from 0 to 2 accordingly. For the  first iteration, it sets the index position to 0, picks up the first element's value and assigns  it to the temporary fruit variable, then executes the indented code block to print out the fruit  as "apple". In the next iteration, the for loop moves the index position to 1, and repeats the  whole process so "berry" is printed out. Once the for loop reaches to the last element and prints  it out as "cherry", it detects that the list is exhausted and marks the whole for loop as ended.
