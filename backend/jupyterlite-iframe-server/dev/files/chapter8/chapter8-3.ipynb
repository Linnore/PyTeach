A string variable can be iterated in the same way except that the indexes are associated with the  characters in the string. In the hello string example, the first element is the character "H" and  the last element is the character "o".

```python
hello = "Hello"
for h in hello:
    print (h)
```

Under certain circumstances, we may want to stop the for loop early or skip executing the indented  code block for some elements. Python supports two control statements for the for loop.

- Break Statement: It brings control out of the loop, and marks the for loop as ended.
- Continue Statement: It ends the execution of the current iteration, returns the control to the  beginning of the loop and starts the next iteration.
