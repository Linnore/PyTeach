The last thing I want to call out is that there are three code blocks in the above example. The statements  in the blue color belong to the main code block. The statements in the green color belong to the indented  for loop code block. The statements in the orange color belong to the indented if code block. In terms of  code execution flow, Python executes the outer code block first, then enters the current statement's child  code block and repeats this process. Once the execution of the deepest inner code block is completed, it  returns the control to the closest outer block. The whole process is very similar to how you open and close  Russian nesting dolls as described at [https://en.wikipedia.org/wiki/Matryoshka_doll](https://en.wikipedia.org/wiki/Matryoshka_doll).
