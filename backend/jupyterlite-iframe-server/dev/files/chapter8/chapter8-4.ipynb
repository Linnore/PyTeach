The change of control flow is normally accompanied with if statement(s), so that the for loop can  be stopped or continued based on the truth of a condition. Here is an example using the break  statement to find the first even number in a list.

```python
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    if num % 2 == 0:
        print (num)
        break
```

The break statement prevents from printing out all the even numbers from the list but the first even  number.
