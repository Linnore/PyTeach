## Chapter 3: Numbers
In the previous lesson, you learned how to manipulate Python "strings" with the addition  and multiplication operators. So what can you do with "numbers" in Python? Actually you can  use Python as a calculator, that supports more operators with numbers than with strings.

- **+**: addition operator, e.g. 2 + 2 = 4
- **-**: subtraction operator, e.g. 10 - 6 = 4
- **\***: multiplication operator, e.g. 5 * 6 = 30
- **/**: division operator, e.g. 8 / 4 = 2

These four operators work exactly the same as common math operators. You can use the print  function to print out the results of these operations.

```python
print(2 + 2)
print(10 - 6)
print(5 * 6)
print(8 / 4)
```

For more complicated operations you can use parentheses ( () ) to group sub expressions.  The resulting equation will follow the same order of operation rules as normal math  operations.

As a quick quiz, can you tell me what will be printed out with below operations using  parentheses?

```python
print(50 - 5 * 6)
print(2 * (4 + 6))
print((10 + 2) / (4 - 1))
```

Besides these four operators, there is one uncommon but very important operator I want  to teach you today. It is called the modulo operator, written as the percentage sign (%).  The modulo operator returns the remainder after dividing one number by another, and it  is a very useful operator in all programming languages.

Equation: 10 % 7 = 3  
Because: 10 / 7 = 1 with a remainder of 3  
Python: "print(10 % 7)" will print out 3  

We can use the modulo operator to calculate the days of a month. For example if the 3rd  day of a month is Wednesday, you will know that the 10th, 17th and 24th days of that  month are also Wednesday because of below facts:

#### Each week has 7 days  
3 % 7 = 3  
10 % 7 = 3  
17 % 7 = 3  
24 % 7 = 3  

Since the remainder is the same for these dates, they must be the same day of the week.

Modulo operator can be used in many other places, for example determining if a number is  odd or even. I will talk more about it in future lessons.

Now let's compare what we learned about strings and about numbers.

Do you still remember what's the key requirement for defining a string? Right, you need  to put the string in a pair of quotes. Try to tell me what's the difference between the  two lines of code below.

```python
print("20")
print (20)
```

Even though the output looks the same, make sure to remember that the two arguments are  different types. Since the first line of code is wrapped by double quotes, it is a string  that contains numbers. On the other hand, the second line doesn't contain quotes, so it is  a number and not a string. The argument in the first line can only use string operators  (+ and *), while the second line can only use number operators (+, -, *, /, %).

Try to execute below lines of code, and find the differences.

```python
print("20" + "20")
print("20" * 2)
print("20" + "students")
print(20 + 20)
print(20 * 2)
```

The string operations in the above example will print "2020" as a string, because it is adding  the content of the string "20" to the first string "20". The results of math operations in the  above example will print 40 as a number. There is a big difference between the string and  number arguments. You always need to be clear on which type of arguments are in use.
