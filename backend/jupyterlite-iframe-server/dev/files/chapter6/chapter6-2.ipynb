### Membership Operators

Besides the comparison operators mentioned above, list variables can also be  manipulated with membership operators that check if a value is a member of  a list or not.

| Operator | Description | Example |
|----|----|----|
| in | Returns True if the specified value is present in the list | variable in list |
| not in | Returns True if the specified value is not present in the list | variable not in list |

Here is a membership operator example. Since "chocolate" is not present in the fruits  list, the found variable is assigned with a False value.

```python
fruits = ["apple", "cherry", "banana"]
favorite = "chocolate"
found = favorite in fruits
print(found)
```

### Boolean Operators

Like the other variables (string, number, list) we have learned, a boolean  variable also has a set of operators that we can use to combine the values  of multiple boolean values.

| Operator | Description | Example |
|----|----|----|
| and | True if both logical conditions are true, otherwise False. | age > 6 and height > 48 |
| or | True if one of the logical conditions is true and False when both logical conditions are false. | age > 8 or height > 52 |
| not | Flips the result, False if the result is true or vice versa. | not found |

Here is a quick look table of the "and" and "or" results with different condition values.  In short, the "and" operator returns True only when both conditions are true, while the  "or" operator returns False only when both conditions are false.

| First Condition | Second Condition | and Result | or Result |
|----|----|----|----|
| True | True | True | True |
| True | False | False | True |
| False | True | False | True |
| False | False | False | False |

Here is a code example on how to use these boolean operators. The is_allowed  variable is assigned with a True value, since the height is larger than 48  even though the age is less than 8. And the not_allowed variable is assigned  with a False value, as it is a reverse of is_allowed.
