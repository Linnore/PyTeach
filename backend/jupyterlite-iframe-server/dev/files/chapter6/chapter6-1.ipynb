### Comparison Operators

The above examples compared variables against strings or numbers, or even  another variable with the same type. Here is a full list of comparison  operators, that can be used with both string and number variables.

| Operator | Description | Example |
|----|----|----|
| == | Equal | a == b |
|!= | Not Equal | a!= b |
| > | Greater than | a > b |
| < | Less than | a < b |
| >= | Greater than or equal to | a >= b |
| <= | Less than or equal to | a <= b |
